```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.isnull(), cmap="viridis", cbar=False, yticklabels=False)
plt.title("Missing Values Heatmap")
plt.show()
```

```python
sns.countplot(x='Survived', data=df)
plt.title("Survival Count")
plt.xlabel("Survived (0 = No, 1 = Yes)")
plt.ylabel("Count")
plt.show()
```

```python
sns.countplot(x='Survived', hue='Sex', data=df)
plt.title("Survival Count by Gender")
plt.show()
```

```python
sns.histplot(df['Age'].dropna(), kde=True, bins=30)
plt.title("Age Distribution")
plt.show()
```

```python
sns.boxplot(x='Pclass', y='Age', data=df)
plt.title("Age Distribution by Passenger Class")
plt.show()
```

```python
sns.pairplot(df[['Survived', 'Pclass', 'Sex', 'Age', 'Fare']], 
             hue='Survived', 
             palette='husl',
             diag_kind='hist')
plt.show()
```

## Summary of Findings

- Out of all passengers, women had a significantly higher survival rate.
- First-class passengers were more likely to survive than second or third-class passengers.
- Younger passengers, especially children, had better survival chances.
- There is a correlation between fare and survival — passengers who paid higher fares had higher survival probability.
- Missing values were mostly found in the 'Age' and 'Cabin' columns.
